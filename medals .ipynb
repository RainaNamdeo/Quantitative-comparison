```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
medals=pd.read_csv("D:/csv/medals_by_country_2016.csv",index_col=0)
medals
```

```python
#line charts
fig,ax=plt.subplots()
ax.plot(medals.index,medals['Bronze'])
plt.show()
```

```python
#for bar chart
fig,ax=plt.subplots()
ax.bar(medals.index,medals['Bronze'])
plt.show()
```

```python
#removing overlapping&adding labels
fig,ax=plt.subplots()
ax.bar(medals.index, medals["Bronze"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_xlabel("countries")
ax.set_ylabel("bronze medal")
ax.set_title("medals by countries")
plt.show()
```

```python
#ading more variable
fig,ax=plt.subplots()
ax.bar(medals.index, medals["Bronze"])
ax.bar(medals.index, medals["Silver"])
ax.set_xlabel("countries")
ax.set_ylabel("bronze medal")
ax.set_title("medals by countries")
ax.set_xticklabels(medals.index, rotation = 45)
plt.show()
```

```python
#overlapping og data
fig,ax=plt.subplots()
ax.bar(medals.index, medals["Bronze"])
ax.bar(medals.index, medals["Silver"], bottom =medals["Bronze"])
ax.set_xticklabels(medals.index, rotation=45)
ax.set_xlabel("countries")
ax.set_ylabel("bronze medal")
ax.set_title("medals by countries")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.bar(medals.index, medals["Bronze"])
ax.bar(medals.index, medals["Silver"], bottom =medals["Bronze"])
ax.bar(medals.index, medals["Gold"], bottom =medals["Silver"] + medals["Bronze"])
ax.set_xticklabels(medals.index, rotation=45)
ax.set_xlabel("countries")
ax.set_ylabel("bronze medal")
ax.set_title("medals by countries")
plt.show()
```
